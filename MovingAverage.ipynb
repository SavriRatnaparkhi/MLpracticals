```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("AirPassengers.csv")
```

```python
data.plot()
plt.show()
```

SMA | Rolling Mean

```python
data['SMA_30'] = data['#Passengers'].rolling(window=30).mean()
```

```python
data.head()
```

```python
data['#Passengers'].plot()
data['SMA_30'].dropna().plot()
plt.show()
print(f'\nConclusion: Passengers column has uptrend / increasing trend (pattern)')
```

CMA

```python
data['CMA'] = data['#Passengers'].expanding().mean()
data.head(5)
```

```python
data['#Passengers'].plot()
data['CMA'].plot()
plt.show()
```

EMA|EWMA

```python
data['EMA'] = data['#Passengers'].ewm(alpha=0.1).mean()
data.sample(5)
```

```python
data['#Passengers'].plot()
data['EMA'].plot()
plt.show()
print(f'Conclusion: As we increease the alpha value, noise in data increased')
```
